```python
from Bio import Phylo
from Bio.Phylo.TreeConstruction import DistanceMatrix, DistanceTreeConstructor
from keyname import keyname as kn
from matplotlib import colors as mpl_colors
from teeplot import teeplot as tp

import Bio
import alifedata_phyloinformatics_convert as apc
import csv
import glob
import itertools
import json
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import networkx as nx
import numpy as np
import pandas as pd
import scipy as sci
import seaborn as sns
import urllib.request as request
import io
from copy import deepcopy


from etlib.SortableTree.SortableTree import SortableTree, find_leaves
from etlib.PlotTools.PlotTools import enable_pretty_graphing, set_color
enable_pretty_graphing()
```

```python
morphs = {x['stint']: x[' morph'][1:] for x in pd.read_csv('morphs_by_stint.csv').to_dict(orient='records')}
```

```python
# store parsimony trees
pars_tree = Phylo.read("data/parsimony_tree.xml", "phyloxml")
sorted_pars_tree = Phylo.read("data/sorted_parsimony_tree.xml", "phyloxml")
rerooted_pars_tree = Phylo.read("data/rerooted_pars_tree.xml", "phyloxml")
```

```python
def name_function(node):
    try:
        return str(int(node.name)) + " (" + morphs[int(node.name)] + ")" if 'Inner' not in str(node) else None
    except TypeError:
        return ""
```

```python
tp.tee(
    Phylo.draw,
    pars_tree,
    label_func=name_function,
    branch_labels=None,
    label_colors=set_color,
    xlabel=["Phylogenetic Distance"],
    ylabel=[""],
    yticks={"ticks": [0]},
    grid={"visible": True, "which": "both", "axis": "x", "color": "gainsboro", "linewidth": "0.5"},
    do_show=False,
    teeplot_subdir="parsimony_tree"
)
```

```python
tp.tee(
    Phylo.draw,
    sorted_pars_tree,
    label_func=name_function,
    branch_labels=None,
    label_colors=set_color,
    xlabel=["Phylogenetic Distance"],
    ylabel=[""],
    yticks={"ticks": [0]},
    grid={"visible": True, "which": "both", "axis": "x", "color": "gainsboro", "linewidth": "0.5"},
    do_show=False,
    teeplot_subdir="sorted_parsimony_tree"
)
```

```python
tp.tee(
    Phylo.draw,
    rerooted_pars_tree,
    label_func=name_function,
    branch_labels=None,
    label_colors=set_color,
    xlabel=["Phylogenetic Distance"],
    ylabel=[""],
    yticks={"ticks": [0]},
    grid={"visible": True, "which": "both", "axis": "x", "color": "gainsboro", "linewidth": "0.5"},
    do_show=False,
    teeplot_subdir="rerooted_parsimony_tree"
)
```

```python
pars_tree_no_outliers = deepcopy(pars_tree)
pars_tree_no_outliers.prune(target=lambda x: int(x.name) == 0)
pars_tree_no_outliers.prune(target=lambda x: int(x.name) == 1)
```
